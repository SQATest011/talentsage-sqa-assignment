Selenium - Browser Setup

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
```

Automate Job Creation

```python
driver.get("https://www.talentsage.com/create-job")

#input fields
input_job_title = driver.find_element(By.ID, "jobtitle")
input_location = driver.find_element(By.ID, "location")
input_description = driver.find_element(By.ID, "desc")

#Enter the data
input_job_title.send_keys("QA Engineer")
input_location.send_keys("New York")
input_description.send_keys("Responsible duties")

#Submit button
button_post = driver.find_element(By.ID, "create")

#call the button
button_post.click()

#element loading
confirmation_message = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.ID, "Confirmation"))
    )

#Assertion - Confirmation
assert "Job created successfully" in confirmation_message.text, "Job creation failed"

driver.quit()
```
